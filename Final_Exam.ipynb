```python
# From Week 2 (SA - 17, 18) - the project plan followed Computational Thinking to divide the task in smaller modules for better anlysis and flexibility
# Create List from Week 5 (2)
# Course constraint: Week 5, 2 concept being used

tasks = []
```

```python
# Adding new items to the tasks/ list
# Course constraint: Week 5, 2, 4 (functions) concept being used

def add_task(task):
    """
    >>> tasks.clear()
    >>> add_task("Pay bills")
    Added: Pay bills
    >>> tasks
    ['Pay bills']
    """

    tasks.append(task) # week 5 (5- modify list)
    print(f"Added: {task}") # Print() Used for debugging as suggested in Week 3/ checking the data insertion
```

```python
# Remove task according to users selection
# index-1 will remove the data from the correct position
# Course constraint: Week 5, 2, 3, 4 (functions) concept being used

def remove_task(index):
    """
    >>> tasks.clear()
    >>> tasks.extend(["A", "B", "C"])
    >>> remove_task(1)
    Removed: A
    >>> tasks
    ['B', 'C']
    >>> remove_task(10)
    Invalid index!
    """

    if index >= 0 and index < len(tasks): # Error 3: IndexError: list index out of range
                                          # Cause of error: Previous condition -> index >= 0 or index < len(tasks)
                                          # the 'or' was allowing the larger inputs to pass, it should be 'and'
                                          # got it becuase of doctest, learned from week 9

        removed = tasks[index-1]
        # del tasks[index-1]
        # tasks.remove("Pay bills") # week 5 (5- modify list)
        tasks.pop(index-1)            # week 5 (5- modify list)
        print(f"Removed: {removed}") # Print() Used for debugging as suggested in Week 3
    else:
        print("Invalid index!")
```

```python
# View all tasks
# Course constraint: Week 2, 3, 5 concept being used

def list_tasks():
    """
    >>> tasks.clear()
    >>> list_tasks()
    No tasks available.

    >>> tasks.extend(["Buy milk", "Study"])
    >>> list_tasks()
    1: Buy milk
    2: Study
    """

    if not tasks:
      print("No tasks available.")
    else:
      for i, t in enumerate(tasks):
        print(f"{i+1}: {t}")
```

```python
# Course constraint: Week 2, 3, 4 concept being used

def display_menu():
    """
    >>> display_menu()   # doctest: +SKIP
    """

    print("\n----------------------")
    print("      Task Manager")
    print("----------------------")
    print("1. Add Task")
    print("2. List Tasks")
    print("3. Remove Task")
    print("4. Quit")
```

```python
# Course constraint: Week 2, 3 (input, print), 4 (function) concept being used

def get_user_choice():
    """
    >>> get_user_choice()   # doctest: +SKIP
    """
    choice = input("Select an option (1-4): ") # Week 3 user input

    if not choice.isdigit(): # checking/ validating input to avoid error, use print statement for the alert message when necessary
        print("Invalid input. Please enter a number between 1 and 4.")
        return -1

    choice = int(choice)

    if choice < 1 or choice > 4:
        print("Please enter a number between 1 and 4.")
        return -1

    return choice
```

```python
# Course constraint: Week 2, 3 (input, print), 4 (function), 5 (while loop) 9 (doctest) concept being used

import doctest # Week 9

def main():
    """
    >>> main()   # doctest: +SKIP
    """

    while True: # week 5 while loop for unknown iteration
        display_menu()
        choice = get_user_choice()

        # Invalid choice handled
        if choice == -1:
            continue

        # OPTION 1: Add Task
        if choice == 1:
            new_task = input("Enter a task (or '#' to go back): ")
            if new_task != "#":
                add_task(new_task)
            continue

        # OPTION 2: List Tasks
        elif choice == 2:
            list_tasks()
            back = input("Enter '#' to return to main menu: ")
            continue   # regardless of input, go back

        # OPTION 3: Remove Task
        elif choice == 3:
            list_tasks()
            if not tasks:
                continue
            data = input("Enter task index number to remove (or '#' to go back): ")

            if data == "#":
                continue

            if not data.isdigit(): # Error 2: ValueError: invalid literal for int() with base 10: 'a'
                                   # The error was caused because there was no validation for the user input, before converting it to integer
                                   # as user want to delete a item from list, they might type the name of the list item instead of writing the index
                                   # checked and validated user input of index
                print("Invalid input. Please enter a valid index.")
                continue

            index = int(data)
            remove_task(index) # Error 1: Solve TypeError: '>=' not supported between instances of 'str' and 'int' (as, in the remove_task function, we are comparing the numbers/integers)
            continue

        # OPTION 4: Quit
        elif choice == 4:
            print("Exiting Task Manager. Goodbye!")
            break


if __name__ == "__main__":
    main()
    doctest.testmod(verbose=True) # Week 9
```
